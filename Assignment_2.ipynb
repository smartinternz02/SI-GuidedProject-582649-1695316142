```python
df["condition of the house"].value_counts()
```

```python
plt.pie(df["condition of the house"].value_counts(),[0.3,0.2,0.2,0,0.2],labels=['3','4','5','2','1'],autopct ='%1.1f%%',colors = ['purple','yellow','blue','green','orange'],)
plt.title('House Condition')
plt.show()
```

```python
df["number of bedrooms"].value_counts()
```

```python
sns.barplot(x =df["number of bedrooms"].value_counts().index,y =df["number of bedrooms"].value_counts() )
```
